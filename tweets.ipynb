```python
import pandas as pd
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np
import re

# Read the CSV file containing the tweets
df = pd.read_csv('data/musk_tweets.csv')

# Filter tweets that contain "Bitcoin" or "Doge"
filtered_df = df[df['Tweet'].str.contains('Bitcoin', case=False, na=False)]

# Join the filtered tweets into a single string
tweets = ' '.join(filtered_df['Tweet'].dropna())

# Replace 'BTC' with 'Bitcoin' and 'DOGE' with 'Doge' for consistency
tweets = tweets.replace('BTC', 'Bitcoin').replace('DOGE', 'Doge')

# Define the colors for the keywords
keywords_colors = {
    'Bitcoin': '#F7931A',
    'Doge': '#BA9F33',
    'Tesla': '#E82127',  # Unused, since we only care about Bitcoin and Doge here
    'SpaceX': '#004B87',  # Unused, since we only care about Bitcoin and Doge here
}

# Count occurrences of 'Bitcoin' and 'Doge' in the filtered tweets
keyword_counts = {keyword: tweets.lower().count(keyword.lower()) for keyword in keywords_colors}

# Initialize word frequencies, starting with the keyword counts
word_frequencies = {}
for word, count in keyword_counts.items():
    word_frequencies[word] = count * 100  # Multiply counts for better visibility

# Add other words to the frequency dictionary from the filtered tweets
for word in re.findall(r'\w+', tweets):
    word_frequencies[word] = word_frequencies.get(word, 0) + 1

# Load the background image (if you have one, such as 'twitter.png')
bg_img = Image.open('twitter.png').convert('L')  # Convert to grayscale
mask = np.array(bg_img)

# Define a color function for the word cloud
def color_func(word, font_size, position, orientation, random_state=None, **kwargs):
    if word in keywords_colors:
        return keywords_colors[word]
    else:
        return '#C0C0C0'  # Default color for other words

# Generate the word cloud using the word frequencies and custom settings
wordcloud = WordCloud(background_color='white',
                      mask=mask,
                      max_words=10000,
                      color_func=color_func,
                      contour_color='#1DA1F2',
                      contour_width=2,
                      relative_scaling=0.5).generate_from_frequencies(word_frequencies)

# Display the word cloud
plt.figure(figsize=(10, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```
